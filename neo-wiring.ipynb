```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000000
R = 4
points = np.random.rand(2,N)*R - R/2
```

```python
# square
#sel = (points[0]<0.5) * (points[1]<0.5) * (points[0]>-0.5) * (points[1]>-0.5)
#circle
#sel = (points[0]**2 + points[1]**2) < 4/np.pi
#ellipse
S = 2
sel = (points[0]**2/(S**2/np.pi) + points[1]**2/(1/np.pi/S**2)) < 1
sel = sel[None].repeat(2,axis=0)
selection = points[sel].reshape(2,-1)
plt.figure(figsize=(10,10))
plt.scatter(points[0], points[1])
plt.scatter(selection[0], selection[1])
```

```python
total_length = np.mean(np.sqrt(selection[0]**2 + selection[1]**2))
area = selection.shape[-1] / N

total_length, area
```
